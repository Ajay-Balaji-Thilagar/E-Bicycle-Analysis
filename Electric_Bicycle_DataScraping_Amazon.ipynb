```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
```

```python
def get_title(soup):
    try:
        title = soup.find("span", attrs = {"id":'productTitle'}).text.strip()
    except AttributeError:
        title = ""
    return title

def get_original_price(soup):
    try:
        original_price = soup.find("span", attrs = {"class" : 'a-price a-text-price apex-basisprice-value'}).text.strip()
        n = len(original_price)
        original_price = original_price[n//2+1:]
    except AttributeError:
        original_price = ""
    return original_price
    
def get_discount_price(soup):
    try:
        discount_price = soup.find("span", attrs = {"class" : 'a-price-whole'}).text.strip()
    except AttributeError:
        discount_price = ""
    return discount_price

def get_rating(soup):
    try:
        rating = soup.find("span",attrs = {"class" : 'a-size-small a-color-base'}).text.strip()
    except AttributeError:
        rating = ""
    return rating

def get_review_count(soup):
    try:
        review_count = soup.find("span", attrs = {"id" : 'acrCustomerReviewText'}).text.strip()[1:-1]
    except AttributeError:
        review_count = ""
    return review_count

def get_full_text(soup):
    
    try:
        
        full_text = soup.get_text(" ", strip=True).lower()
    except:
        
        full_text = ""
        
    return full_text

def get_Battery_Capacity(soup):

    try:

        full_text = get_full_text(soup)

        battery_capacity = re.search(
            r'(\d+(\.\d+)?\s*ah)',
            full_text
        )

        if battery_capacity:

            return battery_capacity.group(1)

        else:

            return np.nan

    except:

        return np.nan

def get_Battery_Life(soup):

    try:

        full_text = get_full_text(soup)

        battery_life = re.search(
            r'(\d+(\.\d+)?\s*year)',
            full_text
        )

        if battery_life:

            return battery_life.group(1)

        else:

            return np.nan

    except:

        return np.nan

def get_Battery_Voltage(soup):

    try:

        full_text = get_full_text(soup)

        battery_voltage = re.search(
            r'(\d+\s*v)',
            full_text
        )

        if battery_voltage:

            return battery_voltage.group(1)

        else:

            return np.nan

    except:

        return np.nan

def get_Battery_Type(soup):

    try:

        full_text = get_full_text(soup)

        battery_type = re.search(
            r'(li[- ]?ion|lithium[- ]?ion)',
            full_text
        )

        if battery_type:

            return battery_type.group(1)

        else:

            return np.nan

    except:

        return np.nan

def get_Charging_Time(soup):

    try:

        full_text = get_full_text(soup)

        charging_time = re.search(
            r'(\d+(\.\d+)?\s*(hrs?|hours?))',
            full_text
        )

        if charging_time:

            return charging_time.group(1)

        else:

            return np.nan

    except:

        return np.nan

def get_Mileage_Per_Charge(soup):

    try:

        full_text = get_full_text(soup)

        mileage = re.search(
            r'((\d+\s*(to|-|~)\s*\d+\s*km[s]?)|(\d+\s*km[s]?))',
            full_text
        )

        if mileage:

            return mileage.group(1)

        else:

            return np.nan

    except:

        return np.nan

def get_Motor(soup):

    try:

        full_text = get_full_text(soup)

        motor = re.search(
            r'(\d+\s*w)',
            full_text
        )

        if motor:

            return motor.group(1)

        else:

            return np.nan

    except:

        return np.nan

def get_Maximum_Speed(soup):

    try:

        full_text = get_full_text(soup)

        maximum_speed = re.search(
            r'(\d+\s*(km/h|kmph|km/hr))',
            full_text
        )

        if maximum_speed:

            return maximum_speed.group(1)

        else:

            return np.nan

    except:

        return np.nan

def get_Weight(soup):

    try:

        full_text = get_full_text(soup)

        weight = re.search(
            r'(\d+(\.\d+)?\s*(kg|kilograms?))',
            full_text
        )

        if weight:

            return weight.group(1)

        else:

            return np.nan

    except:

        return np.nan
```

```python
if __name__ == '__main__':
    base_url = ('https://www.amazon.in/s?k=e+cycle+for+sale&page={}')
    headers = ({"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36", "Accept-Language" : "en-US, en;q=0.5"})
    d = {
    
    "title" : [],
    
    "original_price" : [],
    
    "discount_price" : [],
    
    "rating" : [],
    
    "review_count" : [],
    
    "Battery_Capacity" : [],
    
    "Battery_Life" : [],
    
    "Battery_Voltage" : [],
    
    "Battery_Type" : [],
    
    "Charging_Time" : [],
    
    "Mileage_Per_Charge" : [],
    
    "Motor" : [],
    
    "Maximum_Speed" : [],
        
    "Weight" : []
    }
    
    for pg in range(1,10):
        url = base_url.format(pg)
        webpage = requests.get(url, headers = headers)
        soup = BeautifulSoup(webpage.content, 'html.parser')
        soup = BeautifulSoup(soup.prettify(),'html.parser')
        links = soup.find_all("a", attrs = {'class':'a-link-normal s-line-clamp-2 puis-line-clamp-3-for-col-4-and-8 s-link-style a-text-normal'})
        links_list = []
        
        for link in links:
            links_list.append(link.get('href'))

        for link in links_list:
            new_webpage = requests.get("https://amazon.in" + link, headers=headers)
            new_soup = BeautifulSoup(new_webpage.content, 'html.parser')
            
            d['title'].append(get_title(new_soup))
            d['original_price'].append(get_original_price(new_soup))
            d['discount_price'].append(get_discount_price(new_soup))
            d['rating'].append(get_rating(new_soup))
            d['review_count'].append(get_review_count(new_soup))
            d['Battery_Capacity'].append(get_Battery_Capacity(new_soup))
            d['Battery_Life'].append(get_Battery_Life(new_soup))
            d['Battery_Voltage'].append(get_Battery_Voltage(new_soup))
            d['Battery_Type'].append(get_Battery_Type(new_soup))
            d['Charging_Time'].append(get_Charging_Time(new_soup))
            d['Mileage_Per_Charge'].append(get_Mileage_Per_Charge(new_soup))
            d['Motor'].append(get_Motor(new_soup))
            d['Maximum_Speed'].append(get_Maximum_Speed(new_soup))
            d['Weight'].append(get_Weight(new_soup))
            
        print(f"Completed Page {pg}")
    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df['title'] = amazon_df['title'].replace('', np.nan)
    amazon_df = amazon_df.dropna(subset=['title'])
    amazon_df.to_csv("A:/ajay/AccioJob/Project/amazon_data_raw.csv", header=True, index=False)
    print("CSV File Saved Successfully")
```
